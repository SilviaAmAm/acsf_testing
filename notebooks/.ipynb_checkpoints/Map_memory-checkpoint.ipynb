```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data = np.loadtxt("/Volumes/Transcend/calculations/descriptor_speed/map_memory.txt")
print(data.shape)
```

```python
samples = data[:,0]
memory = data[:,1]
```

```python
new_samples = []
mean_memory = []
stdev_memory = []
iters=10

for i in range(0, len(samples), iters):
    new_samples.append(samples[i])
    mean = np.mean(memory[i:i+iters]) *  9.31e-10
    stdv = np.std(memory[i:i+iters]) *  9.31e-10
    mean_memory.append(mean)
    stdev_memory.append(stdv) 
```

```python
plt.scatter(samples, memory)
plt.show()
```

```python
plt.errorbar(new_samples, mean_memory, yerr=stdev_memory, fmt='o')
plt.xlabel("Number of samples")
plt.ylabel("Virtual Memory (GB)")
plt.show()
```
